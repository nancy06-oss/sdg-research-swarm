```python
import google.generativeai as genai

def researcher_agent(topic: str) -> str:
    """
    AGENT 1 — RESEARCHER
    Role    : Intelligence Gatherer
    Input   : SDG topic (string)
    Output  : Raw research data (string)
    Recovery: Retries with simpler prompt if primary fails
    """
    genai.configure(api_key=API_KEY)
    model = genai.GenerativeModel("gemini-2.5-flash")

    prompt = f"""
    You are a UN SDG Research Agent.

    Topic: {topic}

    Provide detailed information:
    1. Key facts and statistics
    2. Current global challenges
    3. Affected populations
    4. Relevant SDG goals with goal numbers
    5. Current efforts and initiatives

    Think step by step. Use clear headers.
    Be specific with data and numbers.
    """

    print("┌─────────────────────────────────────┐")
    print("│  AGENT 1 — RESEARCHER                │")
    print("│  Status: Gathering intelligence...   │")
    print("└─────────────────────────────────────┘")

    try:
        response = model.generate_content(prompt)
        result   = response.text
        print(f"✅ Agent 1 complete! Output: {len(result)} characters\n")
        return result

    except Exception as e:
        print(f"⚠️  Agent 1 failed: {e}")
        print("🔄 Retrying with fallback prompt...\n")
        fallback = model.generate_content(
            f"Provide basic facts about {topic} and its relation to UN SDGs."
        )
        return fallback.text

# Test it
print("Agent 1 defined ✅")
print("It will run in CELL 9")
```

```python
def synthesizer_agent(raw_research: str, topic: str) -> str:
    """
    AGENT 2 — SYNTHESIZER
    Role    : Pattern Analyst
    Input   : Raw research from Agent 1
    Output  : Structured insights (string)
    Recovery: Uses shortened research if primary fails
    """
    genai.configure(api_key=API_KEY)
    model = genai.GenerativeModel("gemini-2.5-flash")

    prompt = f"""
    You are a UN SDG Synthesis Agent.

    Topic: {topic}

    Raw Research from Agent 1:
    {raw_research}

    Synthesize into structured insights:
    1. TOP 3 Key Insights (most important findings)
    2. Root Causes (why does this problem exist?)
    3. Gaps in current solutions (what is missing?)
    4. Specific SDG targets most relevant (use target numbers like 6.1, 6.2)
    5. Recommended intervention areas (where to act?)

    Think step by step.
    Be analytical and evidence-based.
    """

    print("┌─────────────────────────────────────┐")
    print("│  AGENT 2 — SYNTHESIZER               │")
    print("│  Status: Analyzing insights...       │")
    print("└─────────────────────────────────────┘")

    try:
        response = model.generate_content(prompt)
        result   = response.text
        print(f"✅ Agent 2 complete! Output: {len(result)} characters\n")
        return result

    except Exception as e:
        print(f"⚠️  Agent 2 failed: {e}")
        print("🔄 Using fallback synthesis...\n")
        fallback = model.generate_content(
            f"List the 3 most important points from this research: {raw_research[:500]}"
        )
        return fallback.text

print("Agent 2 defined ✅")
print("It will run in CELL 9")
```

```python
def report_agent(research: str, synthesis: str, topic: str) -> str:
    """
    AGENT 3 — REPORT WRITER
    Role    : Policy Writer
    Input   : Research (Agent 1) + Synthesis (Agent 2)
    Output  : Complete professional SDG report (string)
    Recovery: Generates emergency summary if primary fails
    """
    genai.configure(api_key=API_KEY)
    model = genai.GenerativeModel("gemini-2.5-flash")

    prompt = f"""
    You are a UN SDG Policy Report Agent.

    Topic: {topic}

    Research Data (from Researcher Agent):
    {research}

    Synthesized Insights (from Synthesizer Agent):
    {synthesis}

    Write a complete professional SDG Policy Report
    using EXACTLY this structure:

    # SDG IMPACT REPORT: {topic.upper()}

    ## Executive Summary
    [Write 3 sentences summarizing the issue and urgency]

    ## SDG Alignment
    [List which SDGs apply and exactly why for this topic]

    ## Problem Analysis
    [Evidence-based analysis using the research data]

    ## Key Insights
    [Numbered list from the synthesis — at least 5 points]

    ## Recommended Actions
    [5 concrete, specific, actionable steps]

    ## Success Metrics
    [How to measure impact — be specific with numbers]

    ## Conclusion
    [Forward-looking, inspiring closing statement]

    Make this professional, detailed, and impactful.
    Minimum 600 words.
    """

    print("┌─────────────────────────────────────┐")
    print("│  AGENT 3 — REPORT WRITER             │")
    print("│  Status: Writing policy report...    │")
    print("└─────────────────────────────────────┘")

    try:
        response = model.generate_content(prompt)
        result   = response.text
        print(f"✅ Agent 3 complete! Output: {len(result)} characters\n")
        return result

    except Exception as e:
        print(f"⚠️  Agent 3 failed: {e}")
        print("🔄 Generating emergency summary...\n")
        fallback = model.generate_content(
            f"Write a brief professional SDG policy summary about {topic}."
        )
        return fallback.text

print("Agent 3 defined ✅")
print("It will run in CELL 9")
```

```python
import os
from datetime import datetime

def save_report(topic: str, content: str) -> str:
    """Saves the final report as a .txt file"""
    os.makedirs("/content/outputs", exist_ok=True)
    ts       = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = f"/content/outputs/SDG_{topic.replace(' ', '_')}_{ts}.txt"
    with open(filename, "w", encoding="utf-8") as f:
        f.write(content)
    return filename


def run_swarm(topic: str):
    """
    MASTER ORCHESTRATOR
    Controls the full 3-agent pipeline.
    Includes quality checks and failure recovery between agents.
    """

    start_time = datetime.now()

    print("\n" + "🌍" * 18)
    print("  SDG RESEARCH SWARM — AGENTVERSE ACTIVATED")
    print("🌍" * 18)
    print(f"\n📌 MISSION TOPIC : {topic}")
    print(f"⏰ START TIME    : {start_time.strftime('%H:%M:%S')}")
    print()

    # ── PHASE 1: RESEARCH ──────────────────────────────────
    print("━" * 45)
    print("  PHASE 1 — INTELLIGENCE GATHERING")
    print("━" * 45)
    research = researcher_agent(topic)

    # Quality check
    if not research or len(research) < 50:
        print("⚠️  Research too short — replanning...")
        research = researcher_agent(f"overview of {topic} related to UN SDGs")

    # ── PHASE 2: SYNTHESIS ─────────────────────────────────
    print("━" * 45)
    print("  PHASE 2 — SYNTHESIS & ANALYSIS")
    print("━" * 45)
    synthesis = synthesizer_agent(research, topic)

    # Quality check
    if not synthesis or len(synthesis) < 50:
        print("⚠️  Synthesis too short — using fallback...")
        synthesis = f"Key points extracted: {research[:400]}"

    # ── PHASE 3: REPORT ────────────────────────────────────
    print("━" * 45)
    print("  PHASE 3 — REPORT GENERATION")
    print("━" * 45)
    report = report_agent(research, synthesis, topic)

    # ── PHASE 4: SAVE ──────────────────────────────────────
    print("━" * 45)
    print("  PHASE 4 — SAVING OUTPUT")
    print("━" * 45)
    saved_path = save_report(topic, report)
    print(f"💾 Report saved to: {saved_path}\n")

    # ── SUMMARY ────────────────────────────────────────────
    end_time  = datetime.now()
    duration  = (end_time - start_time).seconds

    print("✅" * 18)
    print("  MISSION COMPLETE")
    print("✅" * 18)
    print(f"\n⏰ END TIME  : {end_time.strftime('%H:%M:%S')}")
    print(f"⌛ DURATION  : {duration} seconds")
    print(f"📊 RESEARCH  : {len(research)} characters")
    print(f"📊 SYNTHESIS : {len(synthesis)} characters")
    print(f"📊 REPORT    : {len(report)} characters")
    print(f"💾 SAVED TO  : {saved_path}")
    print()
    print("=" * 55)
    print("FINAL REPORT:")
    print("=" * 55)
    print(report)

    return report, saved_path

print("Swarm Orchestrator defined ✅")
print("Ready to run in CELL 9!")
```

```python
# ============================================================
#   CHANGE THIS TOPIC TO ANY SDG ISSUE YOU WANT!
# ============================================================

TOPIC = "Clean Water Access in Rural Communities"

# ============================================================
#   RUN — this activates all 3 agents
# ============================================================

final_report, saved_file = run_swarm(TOPIC)
```
